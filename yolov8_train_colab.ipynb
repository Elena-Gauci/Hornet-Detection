```python
data = str(Path("data.yaml").resolve())       # or full path
split = "test"
conf = 0.25
iou = 0.7

W_768 = YOLO("/content/drive/MyDrive/hornet-detection/occlusion-test/runs/yolov8n_768/weights/best.pt")
W_960 = YOLO("/content/drive/MyDrive/hornet-detection/occlusion-test/runs/yolov8n_960/weights/best.pt")

m768 = YOLO(W_768)
m960 = YOLO(W_960)

def eval_model(model, imgsz, tag):
    r = model.val(
        data=data,
        split=split,
        imgsz=imgsz,
        conf=conf,
        iou=iou,
        name=f"compare_test_{tag}",
        verbose=False
    )
    return r

# A) "As-trained" deployment-like evaluation
r768_native = eval_model(m768, 768, "768_native")
r960_native = eval_model(m960, 960, "960_native")

# B) Controlled evaluations (optional but very informative)
r768_at960 = eval_model(m768, 960, "768_at960")
r960_at768 = eval_model(m960, 768, "960_at768")
```

```python
def summarize(r, label):
    # speed keys are usually 'preprocess', 'inference', 'postprocess'
    sp = getattr(r, "speed", {}) or {}
    # metrics object is typically r.box or r.metrics.box in some versions
    box = getattr(r, "box", None) or getattr(getattr(r, "metrics", None), "box", None)

    out = {
        "run": label,
        "mAP50-95": getattr(box, "map", None),
        "mAP50": getattr(box, "map50", None),
        "Precision": getattr(box, "mp", None),
        "Recall": getattr(box, "mr", None),
        "ms_preprocess": sp.get("preprocess", None),
        "ms_inference": sp.get("inference", None),
        "ms_postprocess": sp.get("postprocess", None),
    }
    return out

summary = [
    summarize(r768_native, "768 @768"),
    summarize(r960_native, "960 @960"),
    summarize(r768_at960,  "768 @960"),
    summarize(r960_at768,  "960 @768"),
]

for row in summary:
    print(row)
```

```python
# --- YOLOv8 test-set comparison: 768 vs 960 (overall + per-class) ---
from ultralytics import YOLO
from pathlib import Path
import pandas as pd
import numpy as np
import os
import textwrap

# ---- User inputs (as provided) ----
W_768_PATH = "/content/drive/MyDrive/hornet-detection/occlusion-test/runs/yolov8n_768/weights/best.pt"
W_960_PATH = "/content/drive/MyDrive/hornet-detection/occlusion-test/runs/yolov8n_960/weights/best.pt"
DATA_YAML  = str(Path("data.yaml").resolve())

# ---- Evaluation settings (keep identical for fair comparison) ----
SPLIT = "test"
CONF  = 0.25
IOU   = 0.70

# We'll evaluate each model at:
# 1) its native training imgsz, and
# 2) a controlled shared imgsz (optional but recommended)
EVALS = [
    ("768@768_native", W_768_PATH, 768),
    ("960@960_native", W_960_PATH, 960),
    ("768@960_control", W_768_PATH, 960),
    ("960@768_control", W_960_PATH, 768),
]

# Where to store evaluation artifacts (PR curves, confusion matrices, etc.)
PROJECT_DIR = "/content/drive/MyDrive/hornet-detection/occlusion-test/eval_compare_test"
Path(PROJECT_DIR).mkdir(parents=True, exist_ok=True)

def _safe_float(x):
    try:
        return float(x)
    except Exception:
        return np.nan

def _extract_box_metrics(res):
    """
    Robust extraction across Ultralytics versions:
    - overall: mp (precision), mr (recall), map50, map (map50-95)
    - per-class: p, r, ap50, ap (map50-95)
    """
    # locate the "box" metrics object
    box = getattr(res, "box", None)
    if box is None and hasattr(res, "metrics"):
        box = getattr(res.metrics, "box", None)

    if box is None:
        raise RuntimeError("Could not find box metrics in results object. "
                           "Try printing `type(res)` and `dir(res)`.")

    overall = {
        "Precision": _safe_float(getattr(box, "mp", np.nan)),
        "Recall":    _safe_float(getattr(box, "mr", np.nan)),
        "mAP50":     _safe_float(getattr(box, "map50", np.nan)),
        "mAP50-95":  _safe_float(getattr(box, "map", np.nan)),
    }

    # per-class arrays (names vary slightly across versions)
    # try common attribute names in order
    def get_arr(*names):
        for n in names:
            v = getattr(box, n, None)
            if v is not None:
                return np.array(v, dtype=float)
        return None

    p  = get_arr("p")            # per-class precision
    r  = get_arr("r")            # per-class recall
    ap50 = get_arr("ap50")       # per-class AP@0.50
    ap = get_arr("ap")           # per-class AP@[.50:.95]

    per_class = {"p": p, "r": r, "ap50": ap50, "ap": ap}
    return overall, per_class

def _extract_names(res, fallback=("bee","hornet","other")):
    # Prefer dataset names from results; fallback to your known order.
    names = getattr(res, "names", None)
    if isinstance(names, dict) and len(names) > 0:
        # ensure sorted by class index
        return [names[i] for i in sorted(names.keys())]
    if isinstance(names, list) and len(names) > 0:
        return names
    return list(fallback)

def _extract_speed_ms(res):
    sp = getattr(res, "speed", {}) or {}
    return {
        "ms_preprocess": _safe_float(sp.get("preprocess", np.nan)),
        "ms_inference":  _safe_float(sp.get("inference", np.nan)),
        "ms_postprocess":_safe_float(sp.get("postprocess", np.nan)),
    }

def _find_artifacts(save_dir):
    """
    Ultralytics typically writes:
      - confusion_matrix.png
      - confusion_matrix_normalized.png (sometimes)
      - PR_curve.png
      - P_curve.png, R_curve.png, F1_curve.png
    We'll list the common ones if present.
    """
    p = Path(save_dir)
    candidates = [
        "confusion_matrix.png",
        "confusion_matrix_normalized.png",
        "PR_curve.png",
        "P_curve.png",
        "R_curve.png",
        "F1_curve.png",
        "results.png",
    ]
    found = {c: str(p / c) for c in candidates if (p / c).exists()}
    return found

rows_overall = []
rows_perclass = []
artifact_index = []

for tag, weights_path, imgsz in EVALS:
    model = YOLO(weights_path)

    # Run test evaluation and save plots/artifacts
    res = model.val(
        data=DATA_YAML,
        split=SPLIT,
        imgsz=imgsz,
        conf=CONF,
        iou=IOU,
        project=PROJECT_DIR,
        name=tag,
        plots=True,
        verbose=False,
    )

    # metrics extraction
    overall, per_class = _extract_box_metrics(res)
    speed = _extract_speed_ms(res)
    names = _extract_names(res)

    # overall row
    row = {
        "Run": tag,
        "Weights": weights_path,
        "Eval imgsz": imgsz,
        **overall,
        **speed,
    }
    rows_overall.append(row)

    # per-class rows
    # If per-class arrays are not available, we still print names with NaNs.
    p_arr   = per_class.get("p")
    r_arr   = per_class.get("r")
    ap50_arr= per_class.get("ap50")
    ap_arr  = per_class.get("ap")

    n_cls = len(names)
    for i in range(n_cls):
        rows_perclass.append({
            "Run": tag,
            "Class": names[i],
            "P": _safe_float(p_arr[i]) if p_arr is not None and i < len(p_arr) else np.nan,
            "R": _safe_float(r_arr[i]) if r_arr is not None and i < len(r_arr) else np.nan,
            "AP50": _safe_float(ap50_arr[i]) if ap50_arr is not None and i < len(ap50_arr) else np.nan,
            "AP50-95": _safe_float(ap_arr[i]) if ap_arr is not None and i < len(ap_arr) else np.nan,
        })

    # artifact paths
    save_dir = Path(res.save_dir) if hasattr(res, "save_dir") else Path(PROJECT_DIR) / tag
    found = _find_artifacts(save_dir)
    artifact_index.append({
        "Run": tag,
        "save_dir": str(save_dir),
        **found
    })

# ---- Build polished tables ----
df_overall = pd.DataFrame(rows_overall)

# Make the table easier to read (order + rounding)
col_order = [
    "Run", "Eval imgsz", "Precision", "Recall", "mAP50", "mAP50-95",
    "ms_preprocess", "ms_inference", "ms_postprocess", "Weights"
]
df_overall = df_overall[col_order]

metric_cols = ["Precision","Recall","mAP50","mAP50-95","ms_preprocess","ms_inference","ms_postprocess"]
df_overall[metric_cols] = df_overall[metric_cols].applymap(lambda x: round(float(x), 4) if pd.notnull(x) else x)

df_perclass = pd.DataFrame(rows_perclass)
df_perclass[["P","R","AP50","AP50-95"]] = df_perclass[["P","R","AP50","AP50-95"]].applymap(
    lambda x: round(float(x), 4) if pd.notnull(x) else x
)

# ---- Print tables ----
print("\n=== OVERALL METRICS ON TEST SET (split='test') ===")
display(df_overall)

print("\n=== PER-CLASS METRICS ON TEST SET (split='test') ===")
# Order classes bee, hornet, other if present
class_order = {"bee":0, "hornet":1, "other":2}
df_perclass["ClassOrder"] = df_perclass["Class"].map(lambda c: class_order.get(str(c).lower(), 99))
df_perclass = df_perclass.sort_values(["Run","ClassOrder","Class"]).drop(columns=["ClassOrder"])
display(df_perclass)

# ---- Print exact artifact file paths to cite ----
df_art = pd.DataFrame(artifact_index)

print("\n=== ARTIFACTS (FILES YOU CAN CITE) ===")
for _, r in df_art.iterrows():
    print(f"\nRun: {r['Run']}")
    print(f"  save_dir: {r['save_dir']}")
    for k in ["PR_curve.png","confusion_matrix.png","confusion_matrix_normalized.png","P_curve.png","R_curve.png","F1_curve.png","results.png"]:
        if k in r and isinstance(r[k], str) and len(r[k]) > 0 and os.path.exists(r[k]):
            print(f"  {k}: {r[k]}")
        else:
            # do not spam missing entries; only show if present
            pass

print("\nDone")
```
